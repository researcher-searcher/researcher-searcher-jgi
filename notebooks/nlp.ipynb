```python
import scispacy
import spacy
import pandas as pd
import os
from loguru import logger
from scripts.general import mark_as_complete, load_spacy_model
```

```python
nlp = load_spacy_model()
```

```python
df = pd.read_csv('results/research_data.tsv.gz',sep='\t')
print(df.head())
```

```python
text = df.head(n=100)['title']
docs = list(nlp.pipe(text))
for d in docs[0:10]:
    for sent in d.sents:
        print('\n',sent)
        #print(len(sent.vector))
        print(sent.ents)
        for token in sent:
            print(token.text, token.has_vector, token.vector_norm, token.is_oov)
```

```python
doc1 = docs[0]
doc2 = docs[1]
print(doc1,doc1.vector_norm,'\n',doc2,doc2.vector_norm)
similarity_score=doc1.similarity(doc2)
print(similarity_score)
```
